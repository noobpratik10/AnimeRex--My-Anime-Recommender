```python
# Function to preprocess the Genres column
def preprocess_genres(genres_column):
    genre_mapping = {
        "Slice of Life": "SliceofLife",
        "Avant Garde": "AvantGarde",
        "Award Winning": "AwardWinning",
        "Boys Love": "BoysLove",
        "Girls Love": "GirlsLove",
        "Sci-Fi": "SciFi"
    }

    processed_genres = []
    for genres in genres_column:
        if pd.isnull(genres) or genres.strip() == "UNKNOWN":
            processed_genres.append("")
            continue

        genre_list = genres.split(',')
        processed_genre_list = []
        for genre in genre_list:
            genre = genre.strip()  # Remove leading/trailing whitespace
            genre = genre_mapping.get(genre, genre)  # Replace multi-word genres if needed
            if genre:  # Exclude empty strings
                processed_genre_list.append(genre)

        # Rejoin the processed genres
        processed_genres.append(" ".join(processed_genre_list))

    return processed_genres
```

```python
# Apply the updated preprocessing function
content_df["Genres"] = preprocess_genres(content_df["Genres"])
```

```python
# Replace 'UNKNOWN' with an empty string in the 'Type' column
content_df['Type'] = content_df['Type'].replace('UNKNOWN', '')
```

```python
rating_mapping = {
    "PG-13 - Teens 13 or older": "PGTN",
    "G - All Ages": "G",
    "PG - Children": "PG",
    "R - 17+ (violence & profanity)": "RS",
    "R+ - Mild Nudity": "RP",
    "UNKNOWN": ""
}
content_df['Rating'] = content_df['Rating'].replace(rating_mapping)
print(content_df['Rating'].value_counts())
```

```python
def clean_studios(studio_string):
  if studio_string == 'UNKNOWN':
    return ''

  names = [name.strip() for name in studio_string.split(',')]
  cleaned_names = [name.replace(' ', '').replace('_', '').replace('-','').replace('.','') for name in names]
  return ' '.join(cleaned_names)

# Apply the cleaning function to the specified column
content_df['Studios'] = content_df['Studios'].apply(clean_studios)
```

```python
import re
# Define a preprocessing function for the Synopsis column
def preprocess_synopsis(text):
    if pd.isna(text):
        return ""
    #Remove text after '\n'
    text = text.split('\n', 1)[0]
    #Replace 'No description available for this anime.'
    if text.strip() == 'No description available for this anime.':
        return ""
    #Replace hyphen and long dash
    text = text.replace('-', ' ').replace('—', ' ')
    #Remove numbers
    text = re.sub(r'\d+', '', text)
    #Remove all non-English alphanumeric characters
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)

    return text.strip()
```

```python
content_df['Synopsis'] = content_df['Synopsis'].apply(preprocess_synopsis)
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

#lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# POS tag categories for verbs
VERB_CODES = {'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ'}

def preprocess_tags(text):
    #lowercase
    text = text.lower()
    #Tokenize
    words = nltk.word_tokenize(text)
    #POS tagging
    tags = nltk.pos_tag(words)
    #Process
    processed_words = []
    for i, word in enumerate(words):
        # Lemmatize
        if tags[i][1] in VERB_CODES:
            lemmatized = lemmatizer.lemmatize(word, 'v')
        else:
            lemmatized = lemmatizer.lemmatize(word)
        if lemmatized not in stop_words and lemmatized.isalpha():
            processed_words.append(lemmatized)

    #Join
    final_sent = ' '.join(processed_words)
    #Replace contractions
    contractions = {
        "n't": " not", "'m": " am", "'s": " is",
        "'re": " are", "'ll": " will", "'ve": " have", "'d": " would"
    }
    for contraction, expanded in contractions.items():
        final_sent = final_sent.replace(contraction, expanded)

    return final_sent

content_df['Processed Tags'] = content_df['Tags'].apply(preprocess_tags)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

# Vectorizing using TF-IDF
tfidfvec = TfidfVectorizer(ngram_range=(1, 2), max_features=10000)
tfidf_tags = tfidfvec.fit_transform(content_df["Processed Tags"].fillna(''))

# Using NearestNeighbors
nn = NearestNeighbors(n_neighbors=10, metric='cosine', algorithm='auto')
nn.fit(tfidf_tags)
```

```python
# Function to get k recommendations based on anime_id
def get_recommendations_by_id(anime_id, k=10):
    if anime_id not in content_df['anime_id'].values:
        return f"Anime ID '{anime_id}' not found in the dataset."

    anime_index = content_df[content_df['anime_id'] == anime_id].index[0]
    distances, indices = nn.kneighbors(tfidf_tags[anime_index], n_neighbors=k+1)  # Including the input anime itself
    recommended_indices = indices[0][1:]

    top_k_anime_ids = content_df.iloc[recommended_indices]['anime_id'].values.tolist()

    return top_k_anime_ids
```

```python
def get_anime_id_by_name(anime_name, content_df):
    if anime_name not in content_df['Name'].values:
        return f"Anime '{anime_name}' not found in the dataset."

    anime_id = content_df[content_df['Name'] == anime_name]['anime_id'].values[0]
    return anime_id
```

```python
def print_anime_rows_by_id(anime_ids):
    # Filter the DataFrame based on the anime IDs
    filtered_df = anime_df[anime_df['anime_id'].isin(anime_ids)]
    if not filtered_df.empty:
        print(filtered_df)
    else:
        print(f"No rows found for the provided anime IDs: {', '.join(map(str, anime_ids))}")
```

```python
anime_name=input('Enter the Anime Name Here:')
k=int(input('Enter the no. of recommendations you want:'))

recommended_anime_ids = get_recommendations_by_id(anime_id=get_anime_id_by_name(anime_name,content_df), k=k)

print(f"Top {k} recommended anime IDs for anime ID :",anime_name)
print_anime_rows_by_id(recommended_anime_ids)
```
